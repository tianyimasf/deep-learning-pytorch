```python
import numpy as np
import pandas as pd
import torch
torch.set_printoptions(edgeitems=2, threshold=50, linewidth=75)
```

Data source: https://www.kaggle.com/datasets/albertovidalrod/electricity-consumption-uk-20092022?select=historic_demand_year_2024.csv

```python
electricity_consumption_df = pd.read_csv("../data/p1ch4/electricity_consumption_uk_2009-2024/historic_demand_2009_2024_noNaN.csv") # index_col = 0 so that unnamed index column is not loaded
electricity_consumption_df.head(50)
```

```python
electricity_consumption_df.shape
```

```python
from datetime import datetime

electricity_consumption_df["settlement_year_mon_day"] = electricity_consumption_df["settlement_date"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S').date())
electricity_consumption_df.head()
```

```python
electricity_consumption_groups = electricity_consumption_df.groupby("settlement_year_mon_day")
electricity_consumption_groups
```

Only keep the groups (dates) that has all 48 half hour sessions.

```python
print(f"Initial number of groups: {len(electricity_consumption_groups)}")
electricity_consumption_valid_groups = [group[1] for group in electricity_consumption_groups if group[1].shape[0] == 48]
print(f"Valid number of groups: {len(electricity_consumption_valid_groups)}")
```

Now combine all valid groups together again.

```python
electricity_consumption_df = pd.concat(electricity_consumption_valid_groups, axis=0)
electricity_consumption_df.head()
```

```python
data_cols = electricity_consumption_df.columns[3:19]
electricity_consumption_data = electricity_consumption_df[data_cols]
```

```python
electricity_consumption = torch.from_numpy(electricity_consumption_data.values)
electricity_consumption
```

```python
electricity_consumption.shape, electricity_consumption.stride()
```

```python
daily_electricity_consumption = electricity_consumption.view(-1, 48, electricity_consumption.shape[1])
daily_electricity_consumption.shape, daily_electricity_consumption.stride()
```

```python
daily_electricity_consumption = daily_electricity_consumption.transpose(1, 2)
daily_electricity_consumption.shape, daily_electricity_consumption.stride()
```
